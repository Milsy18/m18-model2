```python
def score_scale(val, y_thresh, g_thresh):
    if val >= g_thresh:
        return 1.0
    elif val >= y_thresh:
        return 0.5
    else:
        return 0.0
```

```python
import numpy as np
import pandas as pd
```

```python
def compute_trend_score(df, market_level_series):
    """
    Adds trendScore to a DataFrame with OHLCV columns and market_level.
    Args:
        df: DataFrame with columns ['Open', 'High', 'Low', 'Close', 'Volume'] and index as Date.
        market_level_series: pd.Series of market level for each date, aligned to df.index.
    Returns:
        trend_scores: pd.Series of trend scores.
    """
    # Compute EMAs (lengths 5, 10, 50, 100, 200)
    ema5   = df['Close'].ewm(span=5, adjust=False).mean()
    ema10  = df['Close'].ewm(span=10, adjust=False).mean()
    ema50  = df['Close'].ewm(span=50, adjust=False).mean()
    ema100 = df['Close'].ewm(span=100, adjust=False).mean()
    ema200 = df['Close'].ewm(span=200, adjust=False).mean()

    # ADX (smoothed directional movement index)
    def compute_adx(high, low, close, length=14):
        up = high.diff()
        down = -low.diff()
        plus_dm = np.where((up > down) & (up > 0), up, 0.0)
        minus_dm = np.where((down > up) & (down > 0), down, 0.0)
        tr1 = high - low
        tr2 = (high - close.shift()).abs()
        tr3 = (low - close.shift()).abs()
        tr = np.maximum.reduce([tr1, tr2, tr3])
        atr = pd.Series(tr).rolling(window=length, min_periods=length).mean()
        plus_di = 100 * pd.Series(plus_dm).rolling(window=length, min_periods=length).sum() / atr
        minus_di = 100 * pd.Series(minus_dm).rolling(window=length, min_periods=length).sum() / atr
        dx = (abs(plus_di - minus_di) / (plus_di + minus_di)) * 100
        adx = dx.rolling(window=length, min_periods=length).mean()
        adx.index = high.index  # Ensure same index
        return adx

    adx_val = compute_adx(df['High'], df['Low'], df['Close'], length=14)

    # Calculate percent distances from EMAs
    ema5_pct = 100 * (df['Close'] - ema5) / ema5
    ema10_pct = 100 * (df['Close'] - ema10) / ema10
    ema50_pct = 100 * (df['Close'] - ema50) / ema50
    ema100_pct = 100 * (df['Close'] - ema100) / ema100
    ema200_pct = 100 * (df['Close'] - ema200) / ema200

    # Trend thresholds
    trend_scores = []
    for idx, row in df.iterrows():
        mkt_level = int(market_level_series.loc[idx]) if idx in market_level_series.index else 5  # fallback
        g_ema5, y_ema5, g_ema10, y_ema10, g_ema50, y_ema50, g_ema100, y_ema100, g_ema200, y_ema200, g_adx, y_adx = get_trend_thresholds(mkt_level)
        t_ema5   = score_scale(ema5_pct.loc[idx],   y_ema5,   g_ema5)
        t_ema10  = score_scale(ema10_pct.loc[idx],  y_ema10,  g_ema10)
        t_ema50  = score_scale(ema50_pct.loc[idx],  y_ema50,  g_ema50)
        t_ema100 = score_scale(ema100_pct.loc[idx], y_ema100, g_ema100)
        t_ema200 = score_scale(ema200_pct.loc[idx], y_ema200, g_ema200)
        t_adx    = score_scale(adx_val.loc[idx],    y_adx,    g_adx)
        # Weights per original: 18%, 17%, 15%, 16%, 14%, 20% (sum=1.00)
        trend_score = (t_ema5 * 0.18 + t_ema10 * 0.17 + t_ema50 * 0.15 +
                       t_ema100 * 0.16 + t_ema200 * 0.14 + t_adx * 0.20) * 30
        trend_scores.append(trend_score)
    return pd.Series(trend_scores, index=df.index, name="trendScore")
```
